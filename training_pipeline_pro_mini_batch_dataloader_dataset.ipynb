```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from torch.utils.data import Dataset, DataLoader
#ok bro
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
df.head()
```

Train_test_split

```python
x_train,x_test,y_train,y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0])
```

Scaling

```python
scalar=StandardScaler()
x_train=scalar.fit_transform(x_train)
x_test=scalar.transform(x_test)
```

```python
x_train
```

Lebel Encoding

```python
y_train
```

```python
encoder=LabelEncoder()
y_train=encoder.fit_transform(y_train)
y_test=encoder.transform(y_test)
```

numpy_to_tensor

```python
x_train_tensor = torch.from_numpy(x_train).float()
x_test_tensor = torch.from_numpy(x_test).float()
y_train_tensor = torch.from_numpy(y_train).float()
y_test_tensor = torch.from_numpy(y_test).float()
```

```python
x_train_tensor.shape
```

```python
y_train_tensor.shape
```

Defining Model

```python
class mysimpleNN(nn.Module):
    def __init__(self,x_num):
      super().__init__()
      self.linear=nn.Linear(x_num,1)
      self.sigmoid=nn.Sigmoid()

    def forward(self, x):
       y_pred=self.linear(x)
       y_pred=self.sigmoid(y_pred)
       return y_pred
```

```python
class Customdataset(Dataset):
  def __init__(self,x,y):
    self.x=x
    self.y=y
  def __len__(self):
    return len(self.x)
  def __getitem__(self,index):
    return self.x[index],self.y[index]
```

```python
# definining sets
#training
train_dataset=Customdataset(x_train_tensor,y_train_tensor)
train_dataloader=DataLoader(train_dataset,batch_size=32,shuffle=True)

#testing
test_dataset=Customdataset(x_test_tensor,y_test_tensor)
test_dataloader=DataLoader(test_dataset,batch_size=32,shuffle=False)
```

parameters

```python
learning_rate = 0.1
epochs = 70
```

```python
# define loss func
loss_function=nn.BCELoss()
```

Training pipeline

```python
#creaating model
model=mysimpleNN(x_train_tensor.shape[1])

#define optimizer
optimizer=torch.optim.SGD(model.parameters(),lr=learning_rate)


for epoch in range(epochs):
  for x_batch,y_batch in train_dataloader:

    #forwardpass
    y_pred=model(x_batch)

    #loss
    loss=loss_function(y_pred,y_batch.view(-1,1))

    #clearing gradient
    optimizer.zero_grad()

    #backwardpass
    loss.backward()

    #parameter update
    optimizer.step()

  #epoch and loss
  print(f"Epoch: {epoch+1}, Loss: {loss.item()}")
```

```python
#evaluation
with torch.no_grad():
  ac=[]
  for x_batch,y_batch in test_dataloader:
    y_pred=model.forward(x_batch)
    y_pred=(y_pred>0.5).float()
    accuracy=(y_pred==y_batch).float().mean()
    ac.append(accuracy)
  print(f"Accuracy: {sum(ac)/len(ac)}")
```
